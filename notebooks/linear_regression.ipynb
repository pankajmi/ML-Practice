```python
# flat
# size(sq ft)       price(lacs)
# 100               10
# 120               11
# 200               20
# 250               24
# 370               35
# 500               45
# 900               89
# 1000              95
# 1400              130

# price = w * size + b

# For following 2 datasets (data1, data2), linear regression isn't working. loss keeps on increasing.
data1 = [[100, 10],
         [120, 11], 
         [200, 20],
         [250, 24],
         [370, 35],
         [500, 45],
         [900, 89],
         [1000, 95],
         [1400, 130]]

data2 = [[100, 119.2],
         [120, 143.2], 
         [200, 239.2],
         [250, 299.2],
         [370, 443.2],
         [120, 299],
         [370, 339.2],
         [200, 120.2],
         [299, 290.3]]


# Fot this dataset it works. Taken from -
# https://developers.google.com/machine-learning/crash-course/linear-regression/gradient-descent
data = [[3.5, 18],
        [3.69, 15],
        [3.44, 18],
        [3.43, 16],
        [4.34, 15],
        [4.42, 14],
        [2.37, 24]]

# data = data2


num_iterations = 1000
# alpha (learning rate)
alpha = .05
w=0
b=0
loss = 0

for i in range(num_iterations):
    # Calculate loss
    
    sum=0
    for d in data:
        # price = w * size + b
        price = w * d[0] + b     
        sum += (d[1]-price)*(d[1]-price)
    loss = sum/len(data)
    
    print('iteration: %d, w: %f, b:%f, loss:%f'%(i, w, b, loss))
    
    # weight slope
    # 1/sample_size*sum(2*size*(price' - price))
    # bias slow
    # 1/sample_size*sum(2 * (price' - price))
    
    # calculate slop with w, b
    dw = 0 
    db = 0
    for d in data:
        size = d[0]
        price = d[1]
        # price = w * size + b
        price1 = w * size + b
        dw += 2 * size * ( price1 - price)
        db += 2 * (price1 - price)
    
    dw = dw / len(data)
    db = db / len(data)
    
    w = w - alpha * dw
    b = b - alpha * db
        
    
```
